```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import category_encoders as ce
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
%matplotlib inline 
```

```python
df = pd.read_csv('healthcare-stroke-data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
bmi_mean = df['bmi'].mean()
df['bmi'].fillna(bmi_mean , inplace = True)
```

```python
encoder = ce.OrdinalEncoder(cols = ['gender','work_type','smoking_status','Residence_type','ever_married'],return_df = True)
df = encoder.fit_transform(df)
```

```python
df.info()
```

```python
df = df.iloc[:,1:]
```

```python
df.head(1)
```

```python
features = df.iloc[: , 0: 10]
label = df.iloc[:,-1]
```

```python
scaled_data = pd.DataFrame(df , columns = ['gender','age','hypertension','heart_disease','ever_married','work_type','Residence_type','avg_glucose_level','bmi','smoking_status'])
```

```python
#train test splitting 

from sklearn.model_selection import train_test_split
feature_train , feature_test , label_train , label_test = train_test_split(features , label ,test_size = 0.2 , random_state = 42)
print('feature_train shape: ', feature_train.shape)
print('label_train shape: ' , label_train.shape)
print('feature_test shape: ', feature_test.shape )
print('label_test shape: ', label_test.shape)
```

```python
#standardscaling and logisticregression model 

from sklearn.preprocessing import StandardScaler 
from sklearn.pipeline import make_pipeline
pipe = make_pipeline(StandardScaler() , LogisticRegression())
pipe.fit(feature_train  , label_train)

```

```python
prediction = pipe.predict(feature_test)
```

```python
result = zip(label_test , prediction)
```

```python
print(list(result))
```

```python
# Evalation of logistic regression 

score = pipe.score(feature_test , label_test)
print(score)
```

```python
# Random Forest Classifier model 

rfc_model = RandomForestClassifier(random_state = 42)
rfc_model.fit(feature_train , label_train )
```

```python
rfc_predictions = rfc_model.predict(feature_test)
result = zip(label_test , rfc_predictions)
print(list(result))
```

```python
# Evaluation of randomforestclassifier model 

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("Accuracy score: " , accuracy_score(label_test , rfc_predictions))
print('\n')
print(classification_report(label_test , rfc_predictions))
```

```python
# Naive Bayes model 

nb_model = GaussianNB()
nb_model.fit(feature_train , label_train)
```

```python
nb_predictions = nb_model.predict(feature_test)
result = zip(label_test , nb_predictions)
print(list(result))
```

```python
# Evaluation of Naive Bayes Model 
from sklearn.metrics import accuracy_score
print("accuracy score : " , accuracy_score(label_test , nb_predictions))
```

```python
knearest_model = KNeighborsClassifier(n_neighbors=3)
knearest_model.fit(feature_train , label_train)
```

```python
knearest_predictions = knearest_model.predict(feature_test)
result = zip(label_test , knearest_predictions)
print(list(result))
```

```python
# Evaluation of Knearest Neigbours 

from sklearn.metrics import accuracy_score 
print("accuracy:" , accuracy_score(label_test , knearest_predictions))
```
