```python
import numpy as np 
import pandas as pd 
import tensorflow as tf 
```

```python
df = pd.read_csv('churn.csv')
df.head()
```

```python
print(f'Shape of dataset: {df.shape[0]} X {df.shape[1]}')
```

```python
#Create a copy of original dataframe.

df1 = df.copy()
```

```python
#Removing CustomerID .

df1.drop('customerID' ,axis = 1, inplace = True)
```

```python
o = (df1.dtypes == 'object')
object_cols = o[o].index 

print(f'Categorical cols: {object_cols}')
```

```python
df1
```

```python
# Converting categorical columns to numeric columns. 

from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()
df1[object_cols]= np.array(oe.fit_transform(df1[object_cols]))
```

```python
#Generating Matrix of Features
x = df1.iloc[:,:-1].values
```

```python
#Generating Dependent Variable Vectors
y = df1.iloc[:,-1].values
```

```python
#Splitting dataset into training and testing dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(f'x_train shape : {x_train.shape}')
print(f'x_test shape :{x_test.shape}')
print(f'y_train shape : {y_train.shape}')
print(f'y_test shape : {y_test.shape}')
```

```python
#Performing Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
#Initialising ANN
ann = tf.keras.models.Sequential()
```

```python
 #Adding Hidden Layers
ann.add(tf.keras.layers.Dense(units=6,activation="relu"))
ann.add(tf.keras.layers.Dense(units=6,activation="relu"))
```

```python
 #Adding Output Layer
ann.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
```

```python
#Compiling ANN
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
```

```python
#Fitting ANN
ann.fit(x_train,y_train,batch_size=32,epochs = 100)
```

```python
ann.evaluate(x_test,y_test)
```

```python
y_test = y_test.reshape(-1,)
y_test[:5]
```

```python
y_pred = ann.predict(x_test)
ann_y_pred_classes = [np.argmax(element) for element in y_pred]
ann_y_pred_classes[:5]
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,ann_y_pred_classes))
```
