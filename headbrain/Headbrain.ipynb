```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

Load the dataset 

```python
data = pd.read_csv(r"headbrain.csv")
data.head()
```

Collecting x and y 

```python
x = data['Head Size(cm^3)'].values 
y = data['Brain Weight(grams)'].values
```

Mean x and y 

```python
mean_x = np.mean(x)
mean_y = np.mean(y)
```

Total number of values 

```python
m = len(x)
```

Using the formula to calculate b1 and b2 

```python
numer = 0 
denom = 0 
for i in range(m):
    numer += (x[i] - mean_x) * (y[i] - mean_y ) 
    denom += (x[i] - mean_x)**2
b1 = numer/denom 
b0 = mean_y - (b1 * mean_x)

#Print Coefficients
print(b0,b1)
```

```python
#Plotting Values and Regression Lines
max_x = np.max(x) + 100 
min_x = np.min(x) - 100 

#calculating line values x and y 
x = np.linspace(min_x,max_x,1000)
y = b0 + b1*x

#Plotting lines 
plt.plot(x,y, color='#58b970' , label='Regression Lines')

#Plotting Scatter Points 
plt.scatter(x,y, c ='#ef5423' , label='Scatter Plot')

plt.xlabel('Head Size in cm^3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

Calculating R^2

```python
sst = 0 
ssr = 0 
for i in range(m):
    y_pred = b0 + b1 *x[i]
    sst += (y[i] - mean_y)**2
    ssr += (y[i] - y_pred)**2
r2 = 1 - (ssr/sst)
print(r2)
```

```python
# Calculating Root Mean Square Error 

rmse = 0 
for i in range(m):
    y_pred = b0 + b1*x[i]
    rmse +=(y[i]-y_pred)**2
rmse = np.sqrt(rmse/m)
print(rmse)
```
