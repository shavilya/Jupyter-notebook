```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import category_encoders as ce 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
%matplotlib inline
```

```python
df = pd.read_csv('machine.csv')
```

```python
df.info()
```

```python
encoder = ce.OrdinalEncoder(cols = ['vendor_name','model'],return_df = True)
transformed = encoder.fit_transform(df)
```

```python
transformed.describe()
```

```python
transformed.info()
```

```python
feature = transformed.iloc[: , :]
label = transformed.iloc[:, 8]
```

```python
feature.drop('prp' , inplace = True , axis = 1)
```

```python
feature_list = list(feature)
label_list = list(label)
print("Its features: ", feature_list)
print("Its label: " , label_list)
```

```python
feature_train , feature_test , label_train ,label_test = train_test_split(feature,label,test_size=0.2,random_state = 42) 
print('shape of feature_train : ',feature_train.shape)
print('shape of feature_test  : ',feature_test.shape  )
print('shape of label_train   : ', label_train.shape)
print('shape of label_test    : ', label_test.shape)
```

```python
model = LinearRegression()
model.fit(feature_train , label_train)
```

```python
pred = model.predict(feature_test)
```

```python
result = zip(label_test , pred )
```

```python

print("RMSE : " , np.sqrt(mean_squared_error(label_test , pred)))
print("R2 Score: " , r2_score(label_test , pred))
```
